```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import matplotlib.pyplot as plt

cotacao_SP500 = web.DataReader('^GSPC', data_source='yahoo', start='2022-01-03', end='2022-05-03')
print(cotacao_SP500)
```

```python
cotacao_SP500['Adj Close'].plot(figsize=(15,5))
```

```python
retorno_SP500 = cotacao_SP500['Adj Close'][-1] / cotacao_SP500['Adj Close'][0]-1
print('Retorno do S&P500 {:.2%}'.format(retorno_SP500))
```

```python
cotacao_SP500['Adj Close'].plot(figsize=(15,5), label = 'S&P500')
cotacao_SP500['Adj Close'].rolling(8).mean().plot(label = 'MM 8')
cotacao_SP500['Adj Close'].rolling(14).mean().plot(label = 'MM 14')
plt.legend()
plt.show()
```
